```python
# Define the bucket capacities and current water levels
bucket_8L = {'capacity': 8, 'current': 8}
bucket_5L = {'capacity': 5, 'current': 0}
bucket_3L = {'capacity': 3, 'current': 0}

def display_buckets():
    """Displays the current state of all buckets."""
    print(f"\n8|{'W' * bucket_8L['current']}{' ' * (bucket_8L['capacity'] - bucket_8L['current'])}|")
    print(" +" + "-" * bucket_8L['capacity'] + "+")
    print(f"5|{'W' * bucket_5L['current']}{' ' * (bucket_5L['capacity'] - bucket_5L['current'])}|")
    print(" +" + "-" * bucket_5L['capacity'] + "+")
    print(f"3|{'W' * bucket_3L['current']}{' ' * (bucket_3L['capacity'] - bucket_3L['current'])}|")
    print(" +" + "-" * bucket_3L['capacity'] + "+")

def fill(bucket):
    """Fill the selected bucket to its full capacity."""
    bucket['current'] = bucket['capacity']

def empty(bucket):
    """Empty the selected bucket."""
    bucket['current'] = 0

def pour(from_bucket, to_bucket):
    """Pour water from one bucket to another until the 'to' bucket is full or 'from' bucket is empty."""
    transfer = min(from_bucket['current'], to_bucket['capacity'] - to_bucket['current'])
    from_bucket['current'] -= transfer
    to_bucket['current'] += transfer

def get_bucket_choice():
    """Prompt user to select a bucket."""
    while True:
        choice = input("Select a bucket (8, 5, or 3) or 'QUIT' to stop: ").lower()
        if choice == '8':
            return bucket_8L
        elif choice == '5':
            return bucket_5L
        elif choice == '3':
            return bucket_3L
        elif choice == 'quit':
            return 'QUIT'
        else:
            print("Invalid choice, please choose again.")

def main():
    print("Water Bucket Puzzle: Try to get exactly 4L of water into one of the buckets!")

    while True:
        display_buckets()

        if bucket_8L['current'] == 4 or bucket_5L['current'] == 4 or bucket_3L['current'] == 4:
            print("\nCongratulations! You have successfully achieved 4L in one of the buckets!")
            break

        action = input("\nWhat would you like to do? (F)ill, (E)mpty, (P)our, or (Q)uit: ").lower()

        if action == 'f':
            selected_bucket = get_bucket_choice()
            if selected_bucket == 'QUIT':
                print("Exiting the game.")
                break
            fill(selected_bucket)

        elif action == 'e':
            selected_bucket = get_bucket_choice()
            if selected_bucket == 'QUIT':
                print("Exiting the game.")
                break
            empty(selected_bucket)

        elif action == 'p':
            print("Choose the bucket to pour from:")
            from_bucket = get_bucket_choice()
            if from_bucket == 'QUIT':
                print("Exiting the game.")
                break
            print("Choose the bucket to pour into:")
            to_bucket = get_bucket_choice()
            if to_bucket == 'QUIT':
                print("Exiting the game.")
                break
            pour(from_bucket, to_bucket)

        elif action == 'q':
            print("Exiting the game.")
            break

        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()

```
